```python
import pandas as pd
from google.colab import drive
drive.mount("/content/drive")
```

```python
# Load the data &inspect data
data = pd.read_csv('/content/drive/MyDrive/Skripsi_Aji Rahman Hakim/Data/Gabungan.csv')
data.head()
```

```python
# prompt: hapus data TimeStamp

data = data.drop('TimeStamp', axis=1)
```

```python
# Count the number of data points for each label
label_counts = data['Label'].value_counts()
print("Label counts before undersampling:\n", label_counts)
```

```python
# Determine the minimum count among the labels for undersampling
min_count = label_counts.min()
```

```python
# Perform undersampling without shuffling the data
undersampled_data = data.groupby('Label').apply(lambda x: x.head(min_count)).reset_index(drop=True)
```

```python
# Display the label counts after undersampling
undersampled_label_counts = undersampled_data['Label'].value_counts()
print("\nLabel counts after undersampling:\n", undersampled_label_counts)
```

```python
# Display the first few rows of the undersampled dataframe
undersampled_data.head()
```

```python
# prompt: hitung data berdasarkan label

# Count the number of data points for each label
label_counts = undersampled_data['Label'].value_counts()

# Print the label counts
print(label_counts)
```

```python
# prompt: unnduh data ke gdrive dengan ekstensi .csv

undersampled_data.to_csv('/content/drive/MyDrive/Skripsi_Aji Rahman Hakim/Data/undersampled_data.csv', index=False)
```
